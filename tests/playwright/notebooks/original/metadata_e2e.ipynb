```python
import os
from pathlib import Path
import qcodes as qc

# Find repo root - try env var first, then search for pyproject.toml
def find_repo_root():
    if "INSTRMCP_REPO_ROOT" in os.environ:
        return Path(os.environ["INSTRMCP_REPO_ROOT"])
    # Search upward from cwd for pyproject.toml
    current = Path.cwd()
    for parent in [current] + list(current.parents):
        if (parent / "pyproject.toml").exists():
            return parent
    raise FileNotFoundError("Could not find repo root (pyproject.toml)")

repo_root = find_repo_root()
station = qc.Station(config_file=str(repo_root / "config" / "station.yaml"))
```
