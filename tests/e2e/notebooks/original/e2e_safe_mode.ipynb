```python
# Setup QCodes mock instruments
import os
from pathlib import Path
import qcodes as qc

def find_repo_root():
    if "INSTRMCP_REPO_ROOT" in os.environ:
        return Path(os.environ["INSTRMCP_REPO_ROOT"])
    current = Path.cwd()
    for parent in [current] + list(current.parents):
        if (parent / "pyproject.toml").exists():
            return parent
    raise FileNotFoundError("Could not find repo root")

repo_root = find_repo_root()
station = qc.Station(config_file=str(repo_root / "config" / "station.yaml"))
```

```python
dac1 = station.load_mock_dac1()
dac2 = station.load_mock_dac2()
```
